Convert Date Time format


```python
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')

df['day_added'] = df['date_added'].dt.day
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year
df['formatted_date'] = df['date_added'].dt.strftime('%d/%m/%Y')

print(df[['date_added', 'formatted_date', 'day_added', 'month_added', 'year_added']])
```

Data Visualization

```python
sns.countplot(df['type'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

Rating of shows and movies

```python
sns.countplot(df['rating'])
plt.xticks(rotation=90)
plt.title('Rating')
plt.show()
```

Relation between Type and Rating

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=df)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
df['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```

Created new genre value on netflix and counted

```python
df_exploded = df['listed_in'].str.split(',').explode().str.strip()
```

```python
genre_counts = df_exploded.value_counts()
```

Process the genres

Plot the genre distribution using Seaborn

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='viridis')
plt.title('Distribution of Movie Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', colors=sns.color_palette('plasma', len(genre_counts)))
plt.title('Distribution of Movie Genres')
```
